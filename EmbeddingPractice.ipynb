## GloVe Model

### Importing The Required Libraries

```python
from torchtext.vocab import GloVe
import torch
import torch.nn as nn
import numpy as np
```

### Importing Model And Model Embeddings

```python
glove_model_6B=GloVe(name="6B",dim=50)
```

```python
glove_embedding=nn.Embedding.from_pretrained(glove_model_6B.vectors,freeze=True)
```

```python
word_to_index=glove_model_6B.stoi
word_to_index["team"]
```

```python
glove_embedding.weight[word_to_index["team"]]
```

### The Input words that has both similar and unsimilar words

```python

words = [
    "taller",
    "short",
    "black",
    "white",
    "dress",
    "pants",
    "big",
    "small",
    "red",
    "blue",
    "smile",
    "frown",
    "race",
    "stroll",
    "tiny",
    "huge",
    "soft",
    "rough",
    "team",
    "individual"
]
```

### Making dictionary of words and their embeddings

```python
dictinory={}
for word in words:
    embed=glove_embedding.weight[word_to_index[word]]
    dictinory[word]=embed
```

### Testing the model and finding words Similar to Small in Sample Space


```python
def find_similar_words(target_word,embeddings,top_k=2):
    similarities={}
    for word,embed in embeddings.items():
        if word!=target_word:
            similar=torch.dot(embed,embeddings[target_word])/(torch.norm(embed)*torch.norm(embeddings[target_word]))
            similarities[word]=similar
    sorted_words = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
    return [word for word,_ in sorted_words[:top_k]]
```

```python
words=find_similar_words("small",dictinory,top_k=2)
words
```
